1. What is hypothesis testing in statistics?
   - Hypothesis testing is a statistical method used to make decisions about a population based on sample data. It helps determine whether there is enough evidence to reject a stated assumption (null hypothesis, H₀) in favor of an alternative hypothesis (H₁).
    - Steps in Hypothesis Testing:
         - a. Define Hypotheses:
           - H₀ (Null Hypothesis): No effect or no difference.
           - H₁ (Alternative Hypothesis): There is an effect or a difference.
         - b. Select Significance Level (α) (e.g., 0.05).
         - c. Choose and Compute the Test Statistic (e.g., t-test, Z-test, ANOVA).
         - d. Find the P-value: Probability of getting the observed result if H₀ is true.
         - Decision Making:
           - If p-value < α, reject H₀ (statistically significant).
           - If p-value ≥ α, fail to reject H₀ (no significant evidence).
   - Example:
      - If testing whether a new drug is more effective than an old one, hypothesis testing helps decide if the observed improvement is due to chance or a real effect.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
   - The null hypothesis (H₀) is a statement that assumes no effect, no difference, or no relationship between variables. It represents the default or status quo assumption.
   - The alternative hypothesis (H₁ or Ha) is a statement that suggests there is an effect, a difference, or a relationship between variables. It challenges the null hypothesis.
   - If p-value < α (e.g., 0.05), we reject H₀ and support H₁. If p-value ≥ α, we fail to reject H₀ (not enough evidence).

3. What is the significance level in hypothesis testing, and why is it important?
   - The significance level (α) in hypothesis testing is the probability of rejecting the null hypothesis (H₀) when it is actually true (Type I error). It represents the threshold for statistical significance.
   - Common Values:
     - 0.05 (5%) → Most common
     - 0.01 (1%) → More strict (reduces false positives)
     - 0.10 (10%) → Less strict (higher chance of detecting an effect)
   - Why is it Important?
     - Controls Type I Error: Limits the risk of falsely concluding an effect exists.
     - Determines Decision Rule: If p-value < α, reject H₀ (significant result).
     - Balances Accuracy and Sensitivity: Lower α reduces false positives but increases false negatives
   - In short, α helps ensure reliable conclusions in hypothesis testing.

4. What does a P-value represent in hypothesis testing?
   - The p-value in hypothesis testing represents the probability of obtaining the observed results (or more extreme results) if the null hypothesis (H₀) is true.
   - Interpretation:
     - p-value < α (e.g., 0.05): Reject the null hypothesis — there is enough evidence to support the alternative hypothesis.
     - p-value ≥ α: Fail to reject the null hypothesis — no significant evidence to support the alternative hypothesis.
   - Summary:
     - A low p-value indicates that the observed data is unlikely under the null hypothesis, suggesting a potential effect or difference. A high p-value suggests the data is consistent with the null hypothesis.

5. How do you interpret the P-value in hypothesis testing?
   - The p-value in hypothesis testing tells you how likely it is to observe the data (or something more extreme) if the null hypothesis (H₀) is true.
   - Interpretation:
     - p-value < α (e.g., 0.05): Reject the null hypothesis — There is enough evidence to support the alternative hypothesis (significant result).
     - p-value ≥ α: Fail to reject the null hypothesis — There is not enough evidence to support the alternative hypothesis (no significant result).
   - In short:
     - Low p-value: Suggests a significant effect (reject H₀).
     - High p-value: Suggests no significant effect (fail to reject H₀).

6. What are Type 1 and Type 2 errors in hypothesis testing?
   - In hypothesis testing, Type 1 and Type 2 errors are potential mistakes in decision-making:
   - Type 1 Error (False Positive):
     - Rejecting the null hypothesis (H₀) when it is actually true.
     - Example: Concluding a drug works when it doesn’t.
   - Type 2 Error (False Negative):
     - Failing to reject the null hypothesis (H₀) when it is actually false.
     - Example: Concluding a drug doesn’t work when it actually does.
   - Summary:
     - Type 1 Error: False alarm (incorrect rejection of H₀).
     - Type 2 Error: Missed detection (incorrect acceptance of H₀).

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
  - In hypothesis testing:
    - One-tailed test:
      - Tests for an effect in only one direction (either greater than or less than).
      - Example: Testing if a drug is greater than a certain value.
    - Two-tailed test:
      - Tests for an effect in both directions (either greater than or less than).
      - Example: Testing if a drug is different from a certain value (could be either higher or lower).
    - Summary:
      - One-tailed: Focuses on one side of the distribution.
      - Two-tailed: Checks both sides for significant differences.

8. What is the Z-test, and when is it used in hypothesis testing?
   - The Z-test is a statistical test used to compare sample data to a population when the population variance is known or the sample size is large (typically n > 30). It is used to determine whether the difference between sample data and the population is statistically significant.
   - When to use the Z-test:
     - Comparing a sample mean to a known population mean (e.g., testing if a sample mean differs from a population mean).
     - Comparing proportions between two groups.
     - Large sample sizes where the Central Limit Theorem applies (n > 30).

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
  - The Z-score measures how many standard deviations a data point is from the mean of a population or sample.
  - Formula to calculate the Z-score:
                           $$Z = \frac{X - \mu}{\sigma}$$
    - X = the data point or sample mean.
    - μ = population mean.
    - σ = population standard deviation.
  - What it represents in hypothesis testing:
    - The Z-score tells you how far your sample statistic (e.g., sample mean) is from the population mean, in terms of standard deviations.
    - A large Z-score (positive or negative) indicates a significant difference from the population mean, suggesting that the result is unlikely to occur by chance.
  - Summary:
    - A Z-score helps assess how extreme or typical a sample value is, and it is used to test hypotheses about population parameters.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
   - The T-distribution is a type of probability distribution used in hypothesis testing, similar to the normal distribution but with heavier tails. It is used when the sample size is small (typically n < 30) and/or when the population standard deviation is unknown.
   - When to use the T-distribution:
     - When the sample size is small (n < 30).
     - When the population standard deviation (σ) is unknown.
     - When the data is approximately normally distributed.
   - Difference from the Normal Distribution:
     - The T-distribution has wider tails to account for more variability with small sample sizes, making it more robust when estimating the population mean.
   - Summary:
     - Use the T-distribution when working with small sample sizes or when the population standard deviation is unknown.

12. What is the T-test, and how is it used in hypothesis testing?

    - Summary:

13. What is the relationship between Z-test and T-test in hypothesis testing?

14. What is a confidence interval, and how is it used to interpret statistical results?
   - A confidence interval is a range of values used to estimate an unknown population parameter (e.g., mean or proportion). It provides a range of plausible values based on sample data, with a specified level of confidence (e.g., 95%).
   - Key Points:
     - A 95% confidence interval means we are 95% confident that the true population parameter lies within the interval.
     - The wider the interval, the less precise the estimate.
     - A narrower interval indicates greater precision.
   - How it's used:
     - If a confidence interval does not include a certain value (e.g., 0 for a difference), it suggests the result is statistically significant.
     - If the interval includes the value (e.g., 0 for a difference), it suggests there is no significant effect.
   - Summary:
     - A confidence interval helps quantify the uncertainty in statistical estimates and gives a range where the true population parameter likely lies.

15. What is the margin of error, and how does it affect the confidence interval?

    - How it affects the confidence interval:
      - The margin of error is added to and subtracted from the sample estimate to create the confidence interval.
      - A larger margin of error results in a wider confidence interval, showing more uncertainty in the estimate.
      - A smaller margin of error results in a narrower confidence interval, showing greater precision in the estimate.
    - Summary:
      - The margin of error affects the width of the confidence interval, with a larger margin leading to less precision and a wider interval

16. How is Bayes' Theorem used in statistics, and what is its significance?

17. What is the Chi-square distribution, and when is it used?
    - The Chi-square distribution is a probability distribution that is used in statistical hypothesis testing, particularly for tests involving categorical data. It is a special case of the gamma distribution and is characterized by its degree of freedom (df).
    - When is it used?
      - Chi-square test for goodness of fit: To test if observed data fits an expected distribution.
      - Chi-square test for independence: To test if two categorical variables are independent.
      - Chi-square test for homogeneity: To test if different populations have the same distribution.
    - Key Characteristics:
      - The distribution is right-skewed and becomes more symmetric as degrees of freedom increase.
      - Used when analyzing frequencies or counts in categories.
    - Summary:
      - The Chi-square distribution is used for hypothesis tests involving categorical data, such as testing relationships or fits between observed and expected frequencies.

18. What is the Chi-square goodness of fit test, and how is it applied?
    - The Chi-square goodness of fit test is a statistical test used to determine if the observed frequency distribution of a categorical variable matches an expected distribution.
    - When is it applied?
      - To test if the observed data fits a specific theoretical distribution (e.g., uniform, normal, or any predefined distribution).
      - Commonly used in situations like testing if dice rolls are fair or if a population follows a specific distribution.
    - Steps in the test:
      - a. State hypotheses:
           - Null hypothesis (H₀): The observed data fits the expected distribution.
           - Alternative hypothesis (H₁): The observed data does not fit the expected distribution.

      - b. Calculate the Chi-square statistic:
           - $\chi^2 = \sum \frac{(O - E)^2}{E}$
           - O = Observed frequency
           - E = Expected frequency
      - c. Compare with the critical value from the Chi-square distribution table using the degrees of freedom (df).
      - d. Decision: If the p-value < α (e.g., 0.05), reject H₀, indicating the observed data does not fit the expected distribution.
   - Summary:
      - The Chi-square goodness of fit test is used to assess whether observed categorical data matches an expected distribution, helping determine the goodness of fit.

19. What is the F-distribution, and when is it used in hypothesis testing?
    - The F-distribution is a probability distribution used in hypothesis testing, particularly when comparing variances between two or more groups.
    - When is it used?
      - ANOVA (Analysis of Variance): To test if there are significant differences in means across multiple groups.
      - F-test: To compare the variances of two populations or groups to see if they are significantly different.
    - Key Characteristics:
      - The F-distribution is right-skewed and its shape depends on the degrees of freedom of the numerator and denominator.
      - It is used when comparing variances or testing the ratio of variances between two populations.
    - Summary:
      - The F-distribution is used to test hypotheses about variances and is commonly applied in ANOVA and F-tests for comparing multiple group variances.

20. What is an ANOVA test, and what are its assumptions?
    - An ANOVA (Analysis of Variance) test is a statistical method used to compare the means of three or more groups to see if at least one group mean is significantly different from the others.
    - Assumptions of ANOVA:
      - Independence: The data points within each group are independent.
      - Normality: The data in each group should be approximately normally distributed.
      - Homogeneity of variance (Equal variances): The variances across groups should be roughly equal.
    - Summary:
      - ANOVA is used to compare means of multiple groups, assuming the data is independent, normally distributed, and has equal variances.

21. What are the different types of ANOVA tests?
    - There are three main types of ANOVA tests:
    - One-Way ANOVA:
      - Compares the means of three or more groups based on one independent variable.
      - Example: Comparing test scores from different teaching methods.
    - Two-Way ANOVA:
      - Compares the means of groups based on two independent variables and checks for interaction between them.
      - Example: Comparing test scores based on both teaching method and gender.
    - Repeated Measures ANOVA:
      - Used when the same subjects are measured multiple times (i.e., within-subjects design).
      - Example: Measuring the weight of individuals at different time points after treatment.
    - Summary:
      - One-way: One factor.
      - Two-way: Two factors (with or without interaction).
      - Repeated measures: Same subjects measured multiple times.

22. What is the F-test, and how does it relate to hypothesis testing?
    - The F-test is a statistical test used to compare the variances of two or more groups to determine if they are significantly different from each other.
    - Relation to Hypothesis Testing:
      - The null hypothesis (H₀): Assumes that the variances of the groups are equal.
      - The alternative hypothesis (H₁): Assumes that at least one group has a different variance.
    - How it works:
      - The F-test calculates the ratio of variances (larger variance / smaller variance). If this ratio is significantly large, it suggests that the variances are different.
    - Use:
      - Often used in ANOVA to compare group variances and assess if there are significant differences in group means.
    - Summary:
      - The F-test is used in hypothesis testing to compare variances, helping to determine if group variances differ significantly.

```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.
#ans.
import numpy as np
from scipy import stats

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    z_score = (np.mean(sample) - pop_mean) / (pop_std / np.sqrt(len(sample)))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
    result = "Reject H₀" if p_value < alpha else "Fail to reject H₀"
    return z_score, p_value, result

# Example
sample = [50, 52, 48, 47, 53, 49, 51, 50, 48, 52]
z, p, conclusion = z_test(sample, pop_mean=50, pop_std=3)

print(f"Z-score: {z:.4f}, P-value: {p:.4f}, Conclusion: {conclusion}")
```

```python
#4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
#ans.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Sample data
np.random.seed(42)
sample = np.random.normal(loc=50, scale=3, size=30)
pop_mean, pop_std, alpha = 50, 3, 0.05

# Z-test
z = (np.mean(sample) - pop_mean) / (pop_std / np.sqrt(len(sample)))
p = 2 * (1 - stats.norm.cdf(abs(z)))
cv = stats.norm.ppf(1 - alpha/2)

# Plot decision regions
x = np.linspace(-4, 4, 1000)
plt.plot(x, stats.norm.pdf(x), label="Standard Normal")
plt.axvline(-cv, color='red', linestyle='dashed')
plt.axvline(cv, color='red', linestyle='dashed')
plt.axvline(z, color='blue', label=f"Z = {z:.2f}")
plt.fill_between(x, stats.norm.pdf(x), where=(x < -cv) | (x > cv), color='red', alpha=0.3)
plt.legend()
plt.show()

print(f"Z-score: {z:.4f}, P-value: {p:.4f}")
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
#ans.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def hypothesis_error_visualization(pop_mean, pop_std, sample_size, effect_size, alpha=0.05):
    x = np.linspace(pop_mean - 4*pop_std, pop_mean + 4*pop_std, 1000)

    # Null (H0) and Alternative (H1) distributions
    h0_dist = stats.norm.pdf(x, pop_mean, pop_std / np.sqrt(sample_size))
    h1_dist = stats.norm.pdf(x, pop_mean + effect_size, pop_std / np.sqrt(sample_size))

    # Critical value for Type 1 error (α)
    critical_value = pop_mean + stats.norm.ppf(1 - alpha) * (pop_std / np.sqrt(sample_size))

    plt.figure(figsize=(8, 5))
    plt.plot(x, h0_dist, label="H0: Null Distribution", color="blue")
    plt.plot(x, h1_dist, label="H1: Alternative Distribution", color="green")
    plt.axvline(critical_value, color='red', linestyle='dashed', label="Critical Value")

    # Shade Type 1 Error (α)
    plt.fill_between(x, h0_dist, where=(x >= critical_value), color='red', alpha=0.3, label="Type 1 Error (α)")

    # Shade Type 2 Error (β)
    beta = stats.norm.cdf(critical_value, loc=pop_mean + effect_size, scale=pop_std / np.sqrt(sample_size))
    plt.fill_between(x, h1_dist, where=(x <= critical_value), color='purple', alpha=0.3, label="Type 2 Error (β)")

    plt.legend()
    plt.title("Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.show()

    print(f"Type 2 Error (β): {beta:.4f}, Power (1-β): {1 - beta:.4f}")

# Example usage
hypothesis_error_visualization(pop_mean=50, pop_std=10, sample_size=30, effect_size=5)
```

```python
#6. Write a Python program to perform an independent T-test and interpret the results.
#ans.
import numpy as np
from scipy import stats

# Simulate two independent samples
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=5, size=30)  # Mean=50, Std=5, Size=30
sample2 = np.random.normal(loc=53, scale=5, size=30)  # Mean=53, Std=5, Size=30

# Perform independent T-test
t_stat, p_value = stats.ttest_ind(sample1, sample2, equal_var=False)

# Print results
print(f"T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
print("Reject H₀" if p_value < 0.05 else "Fail to reject H₀")
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
#ans.
import scipy.stats as stats
import numpy as np

def confidence_interval(data, confidence=0.95):
    # Mean of the sample data
    mean = np.mean(data)

    # Standard error of the sample data
    std_error = np.std(data, ddof=1) / np.sqrt(len(data))

    # Calculate margin of error using the t-distribution
    margin_of_error = stats.t.ppf((1 + confidence) / 2, len(data) - 1) * std_error

    # Confidence interval is mean +/- margin of error
    return mean - margin_of_error, mean + margin_of_error
```

Explanation:
- Confidence interval gives a range of values within which we expect the true population mean to lie.
- It is used to express uncertainty around a sample statistic, such as the sample mean.
- 95% confidence interval means that if you were to take 100 different samples, about 95 of them would contain the true mean.

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
#ans.
def bayes_theorem(prior, likelihood, evidence):
    return (likelihood * prior) / evidence

# Example usage
prior = 0.1  # Probability of hypothesis (before evidence)
likelihood = 0.8  # Probability of evidence given hypothesis
evidence = 0.5  # Total probability of evidence

posterior = bayes_theorem(prior, likelihood, evidence)
print("Posterior Probability:", posterior)
```

- As degrees of freedom increase, it becomes more symmetric and approaches a normal distribution.
- It is used in hypothesis testing (e.g., Chi-square tests).

```python
#16. Implement an F-test using Python to compare the variances of two random samples.
#ans.
import scipy.stats as stats

# Example data: two random samples
sample1 = [23, 29, 21, 25, 30]
sample2 = [31, 33, 32, 29, 30]

# Perform F-test
# stats.f_oneway is used for comparing variances of two or more samples.
# Since we have only two samples here, we can use stats.f.cdf instead.
# Calculate the variances of the two samples.
var1 = np.var(sample1, ddof=1)  # ddof=1 for sample variance
var2 = np.var(sample2, ddof=1)

# Calculate the F-statistic.
f_statistic = var1 / var2

# Calculate the degrees of freedom.
df1 = len(sample1) - 1
df2 = len(sample2) - 1

# Calculate the p-value.
p_value = 2 * min(stats.f.cdf(f_statistic, df1, df2), 1 - stats.f.cdf(f_statistic, df1, df2)) # Two-tailed test


print("F-statistic:", f_statistic)
print("P-value:", p_value)
```

```python
#17. Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results.
#ans.
import scipy.stats as stats

# Sample data for three groups
group1 = [23, 20, 25, 22, 27]
group2 = [30, 31, 29, 35, 33]
group3 = [40, 39, 41, 38, 37]

# Perform ANOVA test
f_stat, p_value = stats.f_oneway(group1, group2, group3)

# Display results
print(f"F-statistic: {f_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: At least one group mean is significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between group means.")
```

```python
#19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.
#ans.
import scipy.stats as stats
import matplotlib.pyplot as plt

def check_anova_assumptions(*groups):
    # Normality test (Shapiro-Wilk)
    for i, group in enumerate(groups, 1):
        stat, p = stats.shapiro(group)
        print(f"Group {i} Normality test: p-value = {p:.4f} {'(Fail)' if p < 0.05 else '(Pass)'}")

    # Homogeneity of variance test (Levene's test)
    stat, p = stats.levene(*groups)
    print(f"Levene's test for equal variance: p-value = {p:.4f} {'(Fail)' if p < 0.05 else '(Pass)'}")

    # Boxplot for visual check
    plt.boxplot(groups, labels=[f'Group {i+1}' for i in range(len(groups))])
    plt.title("Boxplot of Groups")
    plt.ylabel("Values")
    plt.grid(True)
    plt.show()

# Example usage
group1 = [23, 20, 25, 22, 27]
group2 = [30, 31, 29, 35, 33]
group3 = [40, 39, 41, 38, 37]

check_anova_assumptions(group1, group2, group3)
```

Checks performed:
 - Normality → Shapiro-Wilk test (p-value > 0.05 = normal)
 - Equal Variance → Levene’s test (p-value > 0.05 = equal variance)
 - Visual Check → Boxplot for distribution comparison

```python
#24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.
#ans.
import numpy as np
import scipy.stats as stats

# Sample data
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10

# Hypothesized population variance
sigma_sq_hyp = 100  # Expected variance (Std^2 = 10^2)

# Sample variance
s_sq = np.var(data, ddof=1)  # Sample variance (unbiased)

# Compute Chi-square test statistic
n = len(data)
chi_sq_stat = (n - 1) * s_sq / sigma_sq_hyp

# Critical values and p-value
alpha = 0.05
chi_sq_critical_low = stats.chi2.ppf(alpha / 2, df=n - 1)
chi_sq_critical_high = stats.chi2.ppf(1 - alpha / 2, df=n - 1)
p_value = 2 * min(stats.chi2.cdf(chi_sq_stat, df=n - 1), 1 - stats.chi2.cdf(chi_sq_stat, df=n - 1))

# Display results
print(f"Chi-square Statistic: {chi_sq_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
if p_value < alpha:
    print("Reject the null hypothesis: The population variance is significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variance.")
```

```python
#25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.
#ans.
import numpy as np
import statsmodels.api as sm

# Sample data: Success counts and total observations for two groups
success1, total1 = 45, 100  # Group 1: 45 successes out of 100
success2, total2 = 30, 100  # Group 2: 30 successes out of 100

# Proportions
prop1, prop2 = success1 / total1, success2 / total2

# Perform two-proportion Z-test
count = np.array([success1, success2])
nobs = np.array([total1, total2])
z_stat, p_value = sm.stats.proportions_ztest(count, nobs)

# Display results
print(f"Z-statistic: {z_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The proportions are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in proportions.")
```
